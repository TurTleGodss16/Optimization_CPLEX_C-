```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
#Plot API
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y, color = "red");
#plt.style.use("classic");
```

```python
#Object Oriented API
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
fig, ax = plt.subplots(figsize = (5,5));
ax.plot(x, y);
ax.set(title='F(X)',
       xlabel='X-axis',
       ylabel='Y-axis');
```

```python
#Create chart with Plot API
x = np.linspace(0, 10, num = 100);
plt.plot(x, np.sin(x), color='red', label='sinx')
plt.xlabel('X')
plt.ylabel('SinX')
plt.title('F(x)');
#plt.axis([1, 9, -0.5, 1])
plt.plot(x, np.cos(x), color='green', label='cosx');
plt.legend();#print type of label
y = np.linspace(0, 0, num = 100)
plt.plot(x, y, color='blue', label='linear');
#plt.axis()
```
